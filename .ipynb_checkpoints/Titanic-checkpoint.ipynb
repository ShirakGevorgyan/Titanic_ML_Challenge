```python
from sklearn.preprocessing import LabelEncoder
```

```python
X = df_train.drop('Transported', axis=1)
y = df_train['Transported']
```

```python
X_test = df_test.copy()
```

```python
X = X.drop('PassengerId', axis=1)
X_test = X_test.drop('PassengerId', axis=1)
```

```python
for i in X.columns:
    if X[i].dtype == object:
        label = LabelEncoder()
        X[i] = label.fit_transform(X[i])
        X_test[i] = label.transform(X_test[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_val = model.predict(X_val)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred_val))
```

# Best params

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = DecisionTreeClassifier()
```

```python
model_params = {
    "criterion": ["gini", "entropy", "log_loss"],
    "splitter" : ["best", "random"],
    "max_depth": [2, 4, 6, 8],
    "max_features" : [2, 4, 6, "sqrt", "log2"]
    
}
```

```python
finaly_model = GridSearchCV(model,param_grid=model_params, cv=3, verbose=2)
```

```python
finaly_model.fit(X_train, y_train)
```

```python
finaly_model.best_params_
```

```python
y_pred_val = finaly_model.predict(X_val)
```

```python
print(classification_report(y_test, y_pred_val))
```

```python
y_pred = finaly_model.predict(X_test)
```
