```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
df_test = pd.read_csv('data/test.csv')
df_train = pd.read_csv('data/train.csv')
df_sub = pd.read_csv('data/sample_submission.csv')
```

```python
df_train = df_train.drop(['Name', 'Cabin'], axis=1)
df_test = df_test.drop(['Name', 'Cabin'], axis=1)
```

```python
for i in df_train.columns:
    if len(list(df_train[i].value_counts().keys()))<10:
        df_train[i] = df_train[i].apply(lambda x : np.random.choice(a=list(df_train[i].value_counts().keys()), p=[df_train[i].value_counts()[j]/sum(df_train[i].value_counts()) for j in list(df_train[i].value_counts().keys())]) if type(x) == float else x)
    elif len(list(df_train[i].unique())) == len(df_train[i]):
        df_train = df_train.drop(i, axis=1)
    else:
        df_train[i] = df_train[i].interpolate()
```

```python
for i in df_test.columns:
    if len(list(df_test[i].value_counts().keys()))<10:
        df_test[i] = df_test[i].apply(lambda x : np.random.choice(a=list(df_test[i].value_counts().keys()), p=[df_test[i].value_counts()[j]/sum(df_test[i].value_counts()) for j in list(df_test[i].value_counts().keys())]) if type(x) == float else x)
    elif len(list(df_test[i].unique())) == len(df_test[i]):
        df_test = df_test.drop(i, axis=1)
    else:
        df_test[i] = df_test[i].interpolate()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
lable = LabelEncoder()
scaler = StandardScaler()
```

```python
for i in df_test.columns:
    if df_test[i].dtype == object:
        model = LabelEncoder()
        df_test[i]=model.fit_transform(df_test[i])
        df_train[i]=model.fit_transform(df_train[i])
```

```python
new_model = DecisionTreeClassifier()
```

```python
params = {
    'criterion':["gini", "entropy", "log_loss"],
    'splitter':["best", "random"],
    'max_depth':[2, 4],
    'min_samples_split':[3, 5],
    'min_samples_leaf':[1, 2],
    'min_weight_fraction_leaf':[0.0, 1.0],
}
```

```python
full_model= GridSearchCV(new_model , params, cv=3, verbose=2)
```

```python
full_model.fit(X, y)
```

```python
y_pred = full_model.predict(df_test)
```

```python
df_sub['Transported'] = y_pred
```
